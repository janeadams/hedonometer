```python
import pickle
import pandas as pd
from datetime import datetime
import random

import plotly.offline as py
import plotly.graph_objs as go
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
init_notebook_mode(connected=True)
```

```python
mypath = "daily_onegrams/daily_ranks_pickle"
```

```python
from os import listdir
from os.path import isfile, join
wordlist = [f for f in listdir(mypath) if isfile(join(mypath, f))]
```

```python
wordlist
```

```python
compare = ['cyber','activist']
```

```python
traces = []
```

```python
for word in compare:
    df = pd.read_csv(str("daily_onegrams/daily_ranks/"+word+".txt"), sep=" ", skipinitialspace=True)
    df['Date'] = [datetime.strptime(date, '%Y-%m-%d') for date in df['Date']]
    r = lambda: random.randint(0,255)
    color = str('#%02X%02X%02X' % (r(),r(),r()))
    trace = go.Scatter(x=df['Date'], y=df['Rank'], name=str(word),line = dict(color = color))
    traces.append(trace)
data = traces

layout = dict(
    title = str("Comparing "+compare[0]+" to "+compare[1]),
)

fig = dict(data=data, layout=layout)
py.iplot(fig, filename = str(word+'_time-series-simple.html'))
```
