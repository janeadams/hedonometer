```python
import pickle
import pandas as pd
from datetime import datetime
import random

import plotly.offline as py
import plotly.graph_objs as go
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
init_notebook_mode(connected=True)
```

```python
mypath = "daily_onegrams/daily_ranks"
```

```python
from os import listdir
from os.path import isfile, join
wordlist = [f for f in listdir(mypath) if isfile(join(mypath, f))]
```

```python
def load(word):
    #df = pickle.load( open(str("daily_onegrams/daily_ranks_pickle/"+word+".pkl")) )
    df = pd.read_csv(str("daily_onegrams/daily_ranks/"+word+".txt"), sep=" ", skipinitialspace=True)
    df['Date'] = [datetime.strptime(date, '%Y-%m-%d') for date in df['Date']]
    df['Year'] = [date.year for date in df['Date']]
    df['Day'] = [date.timetuple().tm_yday for date in df['Date']]
    viz(df, word)
    
def viz(df, word):    
    r = lambda: random.randint(0,255)
    color = str('#%02X%02X%02X' % (r(),r(),r()))
    trace = go.Scatter(x=df['Date'], y=df['Rank'], name=str(word),line = dict(color = color))
    traces.append(trace)
```

```python
compare = ['digital','cyber','#innovation']
traces = []
for word in compare:
    load(word)
    name = ', '.join(compare)
data = traces
layout = dict(
    title = str("Comparing "+name),
)
fig = dict(data=data, layout=layout)
py.iplot(fig, filename = str(word+'_time-series-simple.html'))
```
